## Problem 1 : Clustering by Living Area and Garage Area

Extract the required variables from the dataset, and then perform Bi-Variate Clustering.     

```python
# Extract the Features from the Data
X = pd.DataFrame(houseData[['GrLivArea','GarageArea']])
                           
# Plot the Raw Data on a 2D grid
f, axes = plt.subplots(1, 1, figsize=(16,8))
plt.scatter(x = "GrLivArea", y = "GarageArea", data = X)                    
```

#### Basic KMeans Clustering

Guess the number of clusters from the 2D plot, and perform KMeans Clustering.    
We will use the `KMeans` clustering model from `sklearn.cluster` module.

```python
# Extract the Features from the Data
X = pd.DataFrame(houseData[['GrLivArea','GarageArea','SalePrice','LotArea', 'TotalBsmtSF']])
logSP = np.log2(houseData['SalePrice']) 
logLotArea = np.log2(houseData['LotArea']) 
X1 = X.copy()
X1['logSP'] = logSP
X1['logLotArea'] = logLotArea
X1 = X1.drop('SalePrice',axis=1) 
X1 = X1.drop('LotArea',axis=1) 

# Plot the Raw Data on 2D grids
sb.pairplot(X1)              
```

```python
X1.head()
```

```python
# Import essential models from sklearn
from sklearn.cluster import KMeans

# Vary the Number of Clusters
min_clust = 1
max_clust = 40

# Compute Within Cluster Sum of Squares
within_ss = []
for num_clust in range(min_clust, max_clust+1):
    kmeans = KMeans(n_clusters = num_clust, n_init=10)
    #kmeans.fit(X)
    kmeans.fit(X1)
    within_ss.append(kmeans.inertia_)

# Angle Plot : Within SS vs Number of Clusters
f, axes = plt.subplots(1, 1, figsize=(16,4))
plt.plot(range(min_clust, max_clust+1), within_ss)
plt.xlabel('Number of Clusters')
plt.ylabel('Within Cluster Sum of Squares')
plt.xticks(np.arange(min_clust, max_clust+1, 1.0))
plt.grid(which='major', axis='y')
plt.show()
```

```python
# Import essential models from sklearn
from sklearn.cluster import KMeans

# Set "optimal" Clustering Parameters
num_clust = 5

# Create Clustering Model using KMeans
kmeans_final = KMeans(n_clusters = num_clust)                 

# Fit the Clustering Model on the Data
#kmeans_final.fit(X)
kmeans_final.fit(X1)
```

```python
# Print the Cluster Centers
print("Features", "\tGLA", "\tGA", "\t\tLA", "\tBSA", "\tSP")
print()

for i, center in enumerate(kmeans_final.cluster_centers_):
    print("Cluster", i, end=":\t")
    for coord in center:
        print(round(coord, 2), end="\t")
    print()
print()

# Print the Within Cluster Sum of Squares
print("Within Cluster Sum of Squares :", kmeans.inertia_)
print()

# Predict the Cluster Labels
#labels = kmeans_final.predict(X)
labels = kmeans_final.predict(X1)

# Append Labels to the Data
X_labeled = X1.copy()
X_labeled["Cluster"] = pd.Categorical(labels)

# Summary of the Cluster Labels
sb.countplot(x=X_labeled["Cluster"])
```

```python
# Plot the Clusters on 2D grids
sb.pairplot(X_labeled, vars = X1.columns.values, hue = "Cluster")
```

```python
# Boxplots for all Features against the Clusters
f, axes = plt.subplots(5, 1, figsize=(16,24))
sb.boxplot(x = 'GrLivArea', y = 'Cluster', data = X_labeled, ax = axes[0])
sb.boxplot(x = 'GarageArea', y = 'Cluster', data = X_labeled, ax = axes[1])
sb.boxplot(x = 'logLotArea', y = 'Cluster', data = X_labeled, ax = axes[2])
sb.boxplot(x = 'TotalBsmtSF', y = 'Cluster', data = X_labeled, ax = axes[3])
#sb.boxplot(x = 'SalePrice', y = 'Cluster', data = X_labeled, ax = axes[4])
sb.boxplot(x = 'logSP', y = 'Cluster', data = X_labeled, ax = axes[4])
```

```python
# Average Behaviour of each Cluster
cluster_data = pd.DataFrame(X_labeled.groupby(by = "Cluster").mean())
cluster_data.plot.bar(figsize = (16,6))
```

## Anomaly Detection for the Dataset

Extract the required variables from the dataset, and then perform Bi-Variate Anomaly Detection.     

```python
# Extract the Features from the Data
X = pd.DataFrame(houseData[['GrLivArea','GarageArea']])
                           
# Plot the Raw Data on a 2D grid
f, axes = plt.subplots(1, 1, figsize=(16,8))
plt.scatter(x = "GrLivArea", y = "GarageArea", data = X)                    
```

#### Basic Anomaly Detection

Use the Nearest Neighbors (k-NN) pattern-identification method for detecting Outliers and Anomalies.    
We will use the `LocalOutlierFactor` neighborhood model from `sklearn.neighbors` module.

```python
# Import LocalOutlierFactor from sklearn.neighbors
from sklearn.neighbors import LocalOutlierFactor

# Set the Parameters for Neighborhood
num_neighbors = 10      # Number of Neighbors
cont_fraction = 0.05   # Fraction of Anomalies

# Create Anomaly Detection Model using LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors = num_neighbors, contamination = cont_fraction)

# Fit the Model on the Data and Predict Anomalies
lof.fit(X)
```

#### Labeling the Anomalies in the Data

We may use the model on the data to `predict` the anomalies.

```python
# Predict the Anomalies
labels = lof.fit_predict(X)

# Append Labels to the Data
X_labeled = X.copy()
X_labeled["Anomaly"] = pd.Categorical(labels)

# Summary of the Anomaly Labels
sb.countplot(x=X_labeled["Anomaly"])
```

```python
# Visualize the Anomalies in the Data
f, axes = plt.subplots(1, 1, figsize=(16,8))
plt.scatter(x = "GrLivArea", y = "GarageArea", c = "Anomaly", cmap = 'viridis', data = X_labeled)
```
