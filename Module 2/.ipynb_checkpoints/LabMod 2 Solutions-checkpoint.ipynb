**e) Plot SalePrice (y-axis) vs LotArea (x-axis) using jointplot, and check the Correlation between the two.**

Extract two variables from the DataFrame -- `SalePrice` and `LotArea` -- and check their mutual relationship.

```python
saleprice = pd.DataFrame(houseDataNum['SalePrice'])
lotarea = pd.DataFrame(houseDataNum['LotArea'])
```

```python
# Set up matplotlib figure with three subplots
f, axes = plt.subplots(nrows=2, ncols=3, figsize=(24, 12))

# Plot the basic uni-variate figures for HP
sb.boxplot(data=saleprice, orient = "h", ax = axes[0,0])
sb.histplot(data=saleprice, kde = False, ax = axes[0,1])
sb.violinplot(data=saleprice, ax = axes[0,2])

# Plot the basic uni-variate figures for Attack
sb.boxplot(data=lotarea, orient = "h", ax = axes[1,0], color = "g")
sb.histplot(data=lotarea, kde = False, ax = axes[1,1], color = "g")
sb.violinplot(data=lotarea, ax = axes[1,2], color = "g")
```

```python
sb.jointplot(x = houseDataNum['LotArea'], y = houseDataNum['SalePrice'], height = 4)
```

```python
# Create a joint dataframe by concatenating the two variables
jointDF = pd.DataFrame(houseDataNum[['LotArea','SalePrice']])

# Calculate the correlation between the two columns/variables
jointDF.corr()
```

```python
sb.heatmap(jointDF.corr(), vmin = -1, vmax = 1, annot = True, fmt=".2f")
```

## Problem 2 : Analysis of Numeric Variables

Extract the required variables from the dataset, as mentioned in the problem.     
`LotArea`, `GrLivArea`, `TotalBsmtSF`, `GarageArea`, `SalePrice`    

```python
houseNumData = pd.DataFrame(houseData[['LotArea', 'GrLivArea', 'TotalBsmtSF', 'GarageArea', 'SalePrice']])
houseNumData.head()
```

**a) Check the individual statistical description and visualize the statistical distributions of each of these variables.**

```python
houseNumData.describe()
```

```python
# Draw the distributions of all variables
f, axes = plt.subplots(5, 3, figsize=(18, 20))
colors = ["r", "g", "b", "m", "c"]

count = 0
for var in houseNumData:
    # During 1st iteration, count = 0 and var = LotArea
    # During 2nd iteration, count = 1 and var = GrLivArea
    sb.boxplot(x=houseNumData[var], orient = "h", color = colors[count], ax = axes[count,0])
    sb.histplot(x=houseNumData[var], color = colors[count], ax = axes[count,1])
    sb.violinplot(x=houseNumData[var], color = colors[count], ax = axes[count,2])
    count += 1
    
plt.show()
```

**b) Comment if the distributions look like “Normal Distribution”, or different. Use the .skew() method to find the
“skewness” of each of the five distributions and verify it with the displots in part a. Which of the variables has the maximum number of outliers?**

```python
houseNumData.skew()
```

**c) Check the relationship amongst the variables using mutual correlation and the correlation heatmap. Comment
which of the variables has the strongest correlation with “SalePrice”. Is this useful in predicting “SalePrice”?**

```python
# Correlation Matrix
print(houseNumData.corr())

# Heatmap of the Correlation Matrix
f, axes = plt.subplots(1, 1, figsize=(10, 10))
sb.heatmap(houseNumData.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
#annot_kws = {"size": 18} defines the font size of annotations. Try 10 to see the effect.
plt.show()
```

**d) Check the relationship amongst the variables using mutual jointplots and an overall pairplot. Comment which of the variables has the strongest linear relation with “SalePrice”. Is this useful in predicting “SalePrice”?**

```python
sb.jointplot(x=houseNumData["LotArea"],y=houseNumData["SalePrice"])
sb.jointplot(x=houseNumData["GrLivArea"],y=houseNumData["SalePrice"])
sb.jointplot(x=houseNumData["TotalBsmtSF"],y=houseNumData["SalePrice"])
sb.jointplot(x=houseNumData["GarageArea"],y=houseNumData["SalePrice"])
plt.show()
```

```python
# Draw pairs of variables against one another
sb.pairplot(data = houseNumData)
plt.show()
```

**Which variables do you think will help us predict `SalePrice` in this dataset?**     

> `GrLivArea` : Possibly the most important variable : Highest Correlation, Strong Linearity    
> `GarageArea` and `TotalBsmtSF` : Important variables : High Correlation, Strong Linearity    
> `LotArea` : Doesn't seem so important as a variable : Low Correlation, Weak Linear Relation
