```python
import pandas as pd
import numpy as np

# 1. Load the dataset
# Assuming 'cars.xlsx' is in the same directory
df = pd.read_excel('cars.xlsx')

# 2. Datatypes of each column
print("--- Column Data Types ---")
print(df.dtypes)
print("\n")

# Comment on Object datatype columns
object_cols = df.select_dtypes(include=['object']).columns
print(f"Number of Object datatype columns: {len(object_cols)}")
print(f"Object Columns: {list(object_cols)}\n")

# 3. Create two dataframes (Object only and Non-Object)
df_object = df.select_dtypes(include=['object'])
df_non_object = df.select_dtypes(exclude=['object'])

# 4. Handle Null (NaN) values in numeric dataframe
# Assuming 'matriculation_id_last2' is the last 2 digits
matriculation_id_last2 = 99 

# Identifying columns with NaN
nan_cols = df.columns[df.isnull().any()].tolist()
print(f"Columns with NaN values: {nan_cols}\n")

# Filling NaN values in the whole dataset
df_filled = df.fillna(value=matriculation_id_last2)

# 5. Using .iloc to find specific row
# Row corresponding to last 2 digits (e.g., index 99)
row_data = df_filled.iloc[matriculation_id_last2]
print(f"--- Data for row {matriculation_id_last2} ---")
print(f"model_name: {row_data['model_name']}")
print(f"year_produced: {row_data['year_produced']}")
print(f"duration_listed: {row_data['duration_listed']}\n")

# 6. Create new dataframe with manufacturer_name and year_produced
df_subset = df[['manufacturer_name', 'year_produced']].copy()

# Sort by year_produced descending
df_sorted = df_subset.sort_values(by='year_produced', ascending=False)
print("--- Sorted DataFrame (Descending Year) ---")
print(df_sorted.head())

# Comment on oldest car manufacturer
oldest_car = df_sorted.iloc[-1]
print(f"\nOldest car manufacturer: {oldest_car['manufacturer_name']} in {oldest_car['year_produced']}")
```
